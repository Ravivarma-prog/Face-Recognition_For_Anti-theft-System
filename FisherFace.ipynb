```python
import cv2
import os
import time
import matplotlib.pyplot as plt
```

```python
classifierFace = cv2.CascadeClassifier("classifiers\\haar\\haarcascade_frontalface_default.xml")
fisherface = cv2.face.FisherFaceRecognizer_create(num_components=50, threshold=1)
fisherface.read("classifiers\\classifierFisherface.yml")

height, width = 220, 220
webcam = cv2.VideoCapture(0, cv2.CAP_DSHOW)

while True:
    isConnected, image = webcam.read()

    if isConnected:
        imageConverted = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        faceDetected = classifierFace.detectMultiScale(imageConverted, scaleFactor=1.5, minSize=(30, 30))

        for (x, y, w, h) in faceDetected:
            start_time = time.time()
            faceImage = cv2.resize(imageConverted[y:y + h, x:x + w], (width, height))
            faceId, accuracy = fisherface.predict(faceImage)
            print("--- %s seconds ---" % (time.time() - start_time))
            faceName = ''

            if faceId == 1 and accuracy < 1000:
                color = (0, 255, 51)
                faceName = 'Nishanth'
            elif faceId == 2 and accuracy < 1000:
                color = (0, 0, 255)
                faceName = 'Bala'
            else:
                color = (0, 0, 255)
                faceName = 'UNKNOWN'

            cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
            cv2.putText(image, f'ID: {faceId}', (x, y + (h + 30)), cv2.FONT_HERSHEY_PLAIN, 1, color)
            cv2.putText(image, f'NAME: {faceName}', (x, y + (h + 50)), cv2.FONT_HERSHEY_PLAIN, 1, color)
            cv2.putText(image, f'ACCURACY: {accuracy}', (x, y + (h + 70)), cv2.FONT_HERSHEY_PLAIN, 1, color)

        cv2.imshow("FACE", image)
        if cv2.waitKey(1) == ord('q'):
            break
        
    
```

```python
def FisherFaceAlgo(image):

    classifierFace = cv2.CascadeClassifier("D:/Phase 2/face_rec/face_rec/fisherface/classifiers/haar/haarcascade_frontalface_default.xml")
    fisherface = cv2.face.FisherFaceRecognizer_create(num_components=50, threshold=1)
    fisherface.read('D:/Phase 2/face_rec/face_rec/fisherface/classifiers/classifierFisherFace.yml')

    height, width = 220, 220
    
    imageConverted = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faceDetected = classifierFace.detectMultiScale(imageConverted, scaleFactor=1.5, minSize=(30, 30))
    
    for (x, y, w, h) in faceDetected:
        start_time = time.time()
        faceImage = cv2.resize(imageConverted[y:y + h, x:x + w], (width, height))
        faceId, accuracy = fisherface.predict(faceImage)
        timetaken = time.time()-start_time
        print("--- %s seconds ---" % (timetaken))
        faceName = ''
        if faceId == 1 and accuracy < 1000:
            color = (0, 255, 51)
            faceName = 'Nishanth'
        elif faceId == 2 and accuracy < 1000:
            color = (0, 0, 255)
            faceName = 'Bala'
        else:
            color = (0, 0, 255)
            faceName = 'UNKNOWN'
        
        
        cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
        cv2.putText(image, f'ID: {faceId}', (x, y + (h + 30)), cv2.FONT_HERSHEY_PLAIN, 1, color)
        cv2.putText(image, f'NAME: {faceName}', (x, y + (h + 50)), cv2.FONT_HERSHEY_PLAIN, 1, color)
        cv2.putText(image, f'ACCURACY: {accuracy}', (x, y + (h + 70)), cv2.FONT_HERSHEY_PLAIN, 1, color)
        
        if faceName == '' or faceName == None:
            faceName = 'Unknown'
        
        return [faceName,timetaken]
        
    
    
    
    
```

```python
import cv2
import os
import numpy

# CLASSIFIERS
eigenface = cv2.face.EigenFaceRecognizer_create(num_components=50)
fisherface = cv2.face.FisherFaceRecognizer_create()

print('STARTED TRAINING...')

def getImageId():
    paths = [os.path.join('D:/Phase 2/face_rec/face_rec/fisherface/images/samples', i) for i in os.listdir('D:/Phase 2/face_rec/face_rec/fisherface/images/samples')]
    faces = []
    ids = []

    for pathImage in paths:
        imageFace = cv2.imread(pathImage)
        imageConverted = cv2.cvtColor(imageFace, cv2.COLOR_BGR2GRAY)
        ids.append(int(os.path.split(pathImage)[-1].split('.')[1]))
        faces.append(imageConverted)
    return numpy.array(ids), faces

ids, faces = getImageId()


fisherface.train(faces, ids)
fisherface.write('D:/Phase 2/face_rec/face_rec/fisherface/classifiers/classifierFisherFace.yml')

print('TRAINING COMPLETED')
```

```python
correct_predictions = 0
ti = 0

for file in test_bala:
    s = 'D:/Phase 2/face_rec/face_rec/fisherface/test/Bala/'+file
    print(s)
    img = cv2.imread(s)
    
    [res,t] = FisherFaceAlgo(img)
    ti += t
    print(res)
    if(res == "Bala"):
        correct_predictions+=1
```
